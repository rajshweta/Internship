```python
# importing necessary libraies

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
# establishing chromedriver connection
driver=webdriver.Chrome(r"C:/Users/nitee/Downloads/chromedriver_win32/chromedriver.exe")

#loading website on browser
driver.get("https://www.naukri.com/")

# filling search element on website with job as Data Analyst and location as Bangalore
search_job=driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys('Data Analyst')
search_loc=driver.find_element_by_id('qsb-location-sugg')
search_loc.send_keys("Bangalore")

search_btn=driver.find_element_by_xpath("//div[@class='search-btn']/button")
search_btn.click()

# defining empty list to store scrapped data
job_titles=[]
companies=[]
locations=[]
experience=[]    
```

```python
# scraping list of job titles
title_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
for i in title_tags:
    title=i.text
    job_titles.append(title)
```

```python
# scraping list of companies
company_tags=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
for i in company_tags:
    company=i.text
    companies.append(company)
```

```python
# scraping list of location
location_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
for i in location_tags:
    location=i.text
    locations.append(location)
```

```python
# scraping list of experience
exp_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span[1]")
for i in exp_tags:
    exp=i.text
    experience.append(exp)
```

```python
# creating dataframe
Jobs=pd.DataFrame()
Jobs['Title']=job_titles[0:10]
Jobs['Company']=companies[0:10]
Jobs['Location']=locations[0:10]
Jobs['Experience']=experience[0:10]
Jobs
```

```python
# establishing chromedriver connection
driver=webdriver.Chrome(r"C:/Users/nitee/Downloads/chromedriver_win32/chromedriver.exe")

#loading website on browser
url="https://www.myntra.com/shoes"
driver.get(url)
time.sleep(1)

# defining empty list to store scrapped data
brands=[]
description=[]
prices=[]

#applying filter on color as black
filter_col=driver.find_element_by_xpath("/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/div")
filter_col.click()
time.sleep(2)

#applying filter on price 
filter_price=driver.find_element_by_xpath("/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/div")
filter_price.click()
time.sleep(2)
```

```python
while(len(brands)<100):
    
    # scraping list of brands
    company_tags=driver.find_elements_by_xpath("//div[@class='product-productMetaInfo']//h3")
    for i in company_tags:
        brand=i.text
        brands.append(brand)
    time.sleep(2)    

    # scraping list of description
    des=driver.find_elements_by_xpath("//div[@class='product-productMetaInfo']//h4[1]")
    for i in des:
        desc=i.text
        description.append(desc)
    time.sleep(2) 

    # scraping list of price
    price_tags=driver.find_elements_by_xpath("//div[@class='product-price']/span[1]")
    for i in price_tags:
        price=i.text
        price=price.split()
        prices.append(price[1])
    time.sleep(2)    

    # switching to next page
    btn=driver.find_element_by_xpath("//a[@rel='next']")
    btn.click()
    time.sleep(2)

    
# creating dataframe
Shoes=pd.DataFrame()
Shoes['Brand']=brands[0:100]
Shoes['Description']=description[0:100]
Shoes['Prices']=prices[0:100]
Shoes
```

```python
# establishing chromedriver connection
driver=webdriver.Chrome(r"C:/Users/nitee/Downloads/chromedriver_win32/chromedriver.exe")

#loading website on browser
driver.get("https://www.ambitionbox.com/")  
```

```python
btn=driver.find_element_by_xpath("//a[@class='link salaries']")
btn.click()
```

```python
# filling search element on website with job as Data Analyst and location as Bangalore
search=driver.find_element_by_xpath("//input[@type='search']")
search.send_keys('Data Scientist')
```

```python
search_btn=driver.find_element_by_xpath("//button[@class='ab_btn search-btn round']")
search_btn.click()
```

```python
# defining empty list to store scrapped data
companies=[]
tsalary=[]
maxsal=[]
avgsal=[]
minsal=[]
exp=[]
```

```python
# scraping list of companies and total salary
title_tags=driver.find_elements_by_xpath("//div[@class='name']")
for i in title_tags:
    title=i.text
    title=title.split("\n")
    companies.append(title[0])
    tsalary.append(title[1])
```

```python
# scraping list of min and max salary
tags=driver.find_elements_by_xpath("//div[@class='value body-medium']")
for i in tags:
    mins=i.text
    minsal.append(mins)
```

```python
# scraping list of avg salary
tags=driver.find_elements_by_xpath("//p[@class='averageCtc']")
for i in tags:
    avg=i.text
    avgsal.append(avg)
```

```python
# scraping list of avg salary
tags=driver.find_elements_by_xpath("//div[@class='salaries sbold-list-header']")
for i in tags:
    ex=i.text
    ex=ex.split("\n")
    exp.append(ex[2])
```

```python
# creating dataframe
Jobs=pd.DataFrame()
Jobs['Companies']=companies[0:10]
Jobs['Total salaries']=tsalary[0:10]
Jobs['Min salary']=minsal[0::2]
Jobs['Avg salary']=avgsal[0:10]
Jobs['Max Salary']=minsal[1::2]
Jobs['Experience']=exp[0:10]
Jobs
```

```python
# closing the driver
driver.close()
```
